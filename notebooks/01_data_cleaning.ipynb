```python
import pandas as pd
import numpy as np
```

```python
path = r"E:\E-commerce Analysis\Data\raw\fato_vendas.csv"
```

```python
df_FatoVendas = pd.read_csv (path, sep= ';')
```

```python
df_FatoVendas.info()
```

```python
df_FatoVendas.head()
```

```python
df_FatoVendas['order_purchase_timestamp'] = df_FatoVendas['order_purchase_timestamp'].astype('datetime64[ns]')
df_FatoVendas['order_status'] = df_FatoVendas['order_status'].astype('category')
df_FatoVendas['customer_state'] = df_FatoVendas['customer_state'].astype('category')
df_FatoVendas['product_category_name'] = df_FatoVendas['product_category_name'].astype('category')

df_FatoVendas['price'] = df_FatoVendas['price'].astype('float32')
df_FatoVendas['freight_value'] = df_FatoVendas['freight_value'].astype('float32')
df_FatoVendas['total_payment'] = df_FatoVendas['total_payment'].astype('float32')
```

```python
df_FatoVendas[df_FatoVendas.isnull().any(axis=1)]
```

```python
df_FatoVendas[
    (df_FatoVendas['order_status'].notnull()) &
    (df_FatoVendas['product_category_name'].isnull())
]
```

```python
df_FatoVendas['product_category_name'] = df_FatoVendas['product_category_name'].cat.add_categories(['unknown'])

df_FatoVendas.loc[
    (df_FatoVendas['order_status'] == 'delivered') &
    (df_FatoVendas['product_category_name'].isnull()),
    'product_category_name'
] = 'unknown'
```

```python
df_FatoVendas.isnull().sum()
```

```python
df_FatoVendas['product_category_name'] = df_FatoVendas['product_category_name'].fillna('unknown')
```
