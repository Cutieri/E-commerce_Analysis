```python
import pandas as pd
import numpy as np
```

```python
path = r'E:\E-commerce Analysis\Data\raw\perfil_cliente.csv'
df = pd.read_csv(path, sep=';', encoding='cp1252')
```

```python
df['total_gasto'] = df['total_gasto'].replace(',', '.')
df['total_gasto'] = df['total_gasto'].round(2)
```

```python
df.head()
```

```python
#Quantos clientes compraram mais de 1 vez
clientes_fieis = df[df['numero_pedidos'] > 1]
print(f'Número de clientes fiéis: {clientes_fieis.shape[0]}')
```

```python
#Quantidade Cliente não recorrentes
clientes_nao_fieis = df[df['numero_pedidos'] == 1]
print(f'Número de clientes não fiéis: {clientes_nao_fieis.shape[0]}')
```

```python
#Taxa de recorrencia
taxa_recorrencia = (clientes_fieis.shape[0] / df.shape[0]) * 100
print(f'Taxa de recorrência: {taxa_recorrencia:.2f}%')  
```
