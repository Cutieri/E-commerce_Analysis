```python
import pandas as pd
```

```python
path = r'E:\E-commerce Analysis\Data\raw\performance_produto.csv'
```

```python
df = pd.read_csv(path, sep=';')
```

```python
df.head(10)
```

```python
df.isnull().info()
```

```python
df['product_id'] = df['product_id'].astype('str')
df['product_category_name'] = df['product_category_name'].astype('category')
df['quantidade_vendida'] = df['quantidade_vendida'].astype('int')
df['receita_total'] = df['receita_total'].astype('float')
df.info()
```

```python
df['product_category_name'] = df['product_category_name'].cat.add_categories(['unknown'])
df['product_category_name'] = df['product_category_name'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
df.to_csv("performance_produto_tratada.csv", index=False)
```
